```python
# استيراد المكاتب اللي هنحتاجها
import pandas as pd
import numpy as np
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import accuracy
import matplotlib.pyplot as plt

# 1. قراءة الداتا من ملف Excel
# بنقرأ الملف بتاعك، لو عندك ملف جديد (زي egypt_tourism_data_5000.xlsx) غيّر الاسم
data = pd.read_excel("tourism_behavior_analysis_updated_data.xlsx")

# 2. تحضير الداتا لـ Surprise
# بنختار الأعمدة المهمة: Tourist_ID (المستخدم)، Preferred_Destination (العنصر)، Spending_USD (التقييم)
interactions = data[['Tourist_ID', 'Preferred_Destination', 'Spending_USD']]

# 3. تحديد نطاق التقييمات (Rating Scale)
# بنحدد إن الـ Spending_USD بيتراوح من أقل قيمة لأعلى قيمة في الداتا
reader = Reader(rating_scale=(interactions['Spending_USD'].min(), interactions['Spending_USD'].max()))

# 4. تحميل الداتا في Surprise
# بنحوّل البيانات لشكل Surprise يفهمه
dataset = Dataset.load_from_df(interactions[['Tourist_ID', 'Preferred_Destination', 'Spending_USD']], reader)

# 5. تقسيم الداتا للتدريب والاختبار
# بنقسم الداتا: 80% تدريب، 20% اختبار
trainset, testset = train_test_split(dataset, test_size=0.2)

# 6. إنشاء نموذج SVD
# SVD هو خوارزمية قوية للتوصية تعتمد على تحليل المصفوفات
model = SVD(n_factors=50, n_epochs=20, lr_all=0.005, reg_all=0.02)

# 7. تدريب النموذج
# بنمرن النموذج على داتا التدريب
model.fit(trainset)

# 8. اختبار النموذج
# بنقيّم النموذج على داتا الاختبار باستخدام RMSE
predictions = model.test(testset)
rmse = accuracy.rmse(predictions)
print(f"RMSE: {rmse:.4f}")

# 9. دالة للتوصية لمستخدم معين
def recommend_destinations(user_id, model, data, top_n=5):
    # نجيب كل الوجهات اللي في الداتا
    all_destinations = data['Preferred_Destination'].unique()
    # نجيب الوجهات اللي السائح زارها فعلاً
    visited = data[data['Tourist_ID'] == user_id]['Preferred_Destination'].values
    # نستبعد الوجهات اللي زارها من القايمة
    to_predict = [dest for dest in all_destinations if dest not in visited]
    # نعمل توقعات لكل وجهة
    predictions = [model.predict(user_id, dest) for dest in to_predict]
    # نرتب التوقعات من الأعلى للأقل بناءً على التقييم المتوقع
    predictions.sort(key=lambda x: x.est, reverse=True)
    # نرجع أفضل top_n وجهات
    top_destinations = [pred.iid for pred in predictions[:top_n]]
    return top_destinations

# 10. اختبار التوصية لسائح معين (مثلاً T0001)
user_id = 'T0001'
recommendations = recommend_destinations(user_id, model, data)
print(f"التوصيات للسائح {user_id}: {recommendations}")

# 11. رسم بياني للتقييمات المتوقعة (اختياري)
# بنرسم التقييمات المتوقعة لـ T0001 لكل الوجهات
all_destinations = data['Preferred_Destination'].unique()
scores = [model.predict(user_id, dest).est for dest in all_destinations]
plt.bar(np.arange(len(scores)), scores)
plt.xlabel("Destination Index")
plt.ylabel("Predicted Spending (USD)")
plt.title(f"Predicted Scores for {user_id}")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

# 1. قراءة الداتا
data = pd.read_excel(r"D:\project_movie\data\Code\egypt_tourism_data_5000.xlsx")

# 2. تحضير الداتا
# بنعمل مصفوفة (pivot table) للسياح والوجهات، والقيم هي Spending_USD
pivot_table = data.pivot_table(index='Tourist_ID', columns='Preferred_Destination', values='Spending_USD', fill_value=0)

# 3. حساب التشابه بين السياح
# بنستخدم Cosine Similarity عشان نشوف مين قريب من مين
similarity_matrix = cosine_similarity(pivot_table)

# 4. دالة للتوصية
def recommend_destinations(user_id, pivot_table, similarity_matrix, top_n=5):
    # نجيب رقم السائح في المصفوفة
    user_idx = pivot_table.index.get_loc(user_id)
    # نجيب السياح اللي شبهه (أعلى تشابه)
    similar_users = similarity_matrix[user_idx].argsort()[::-1][1:]  # نستبعد السائح نفسه
    # نجيب الوجهات اللي زارها السياح الشبهه
    similar_users_ratings = pivot_table.iloc[similar_users]
    # نجمع الإنفاق لكل وجهة ونرتبها
    destination_scores = similar_users_ratings.mean(axis=0).sort_values(ascending=False)
    # نستبعد الوجهات اللي السائح زارها فعلاً
    visited = data[data['Tourist_ID'] == user_id]['Preferred_Destination'].values
    recommendations = [dest for dest in destination_scores.index if dest not in visited]
    # نرجع أفضل top_n وجهات
    return recommendations[:top_n]

# 5. اختبار التوصية
user_id = 'T0001'
recommendations = recommend_destinations(user_id, pivot_table, similarity_matrix)
print(f"التوصيات للسائح {user_id}: {recommendations}")

# 6. رسم بياني (اختياري)
user_idx = pivot_table.index.get_loc(user_id)
scores = pivot_table.iloc[user_idx]
plt.bar(range(len(scores)), scores)
plt.xlabel("Destination Index")
plt.ylabel("Spending (USD)")
plt.title(f"Spending for {user_id}")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# 1. قراءة الداتا من الملف
data = pd.read_excel(r"D:\project_movie\data\Code\egypt_tourism_data_5000.xlsx")

# 2. تحضير الداتا
# نختار الأعمدة المهمة لنظام Content-Based (الوجهات وإنفاق السياح)
features = data[['Preferred_Destination', 'Spending_USD', 'Accommodation_Type']]

# 3. تحويل البيانات لشكل عددي
# نستخدم One-Hot Encoding لتحويل Accommodation_Type لأرقام
features_encoded = pd.get_dummies(features, columns=['Accommodation_Type'], prefix='Accommodation')

# 4. تجميع البيانات حسب الوجهة (متوسط الإنفاق لكل وجهة)
destination_features = features_encoded.groupby('Preferred_Destination').mean().reset_index()

# 5. حساب مصفوفة التشابه بين الوجهات
# نستبعد عمود Preferred_Destination لأنه نص مش رقمي
feature_matrix = destination_features.drop('Preferred_Destination', axis=1)
similarity_matrix = cosine_similarity(feature_matrix)

# 6. دالة التوصية
def recommend_destinations(tourist_id, data, destination_features, similarity_matrix, top_n=3):
    # نجيب الوجهة المفضلة للسائح
    preferred_dest = data[data['Tourist_ID'] == tourist_id]['Preferred_Destination'].values[0]
    
    # نلاقي رقم الوجهة في المصفوفة
    dest_idx = destination_features[destination_features['Preferred_Destination'] == preferred_dest].index[0]
    
    # نجيب التشابه بتاع الوجهة دي مع كل الوجهات التانية
    similar_scores = similarity_matrix[dest_idx]
    
    # نرتب الوجهات حسب التشابه (من الأعلى للأقل)
    similar_indices = similar_scores.argsort()[::-1]
    
    # نستبعد الوجهة نفسها ونجيب أعلى top_n وجهات
    recommendations = [
        destination_features.iloc[idx]['Preferred_Destination'] 
        for idx in similar_indices if idx != dest_idx
    ][:top_n]
    
    return recommendations

# 7. اختبار التوصية
tourist_id = 'T0001'
recommendations = recommend_destinations(tourist_id, data, destination_features, similarity_matrix)
print(f"التوصيات للسائح {tourist_id} (الوجهة المفضلة: {data[data['Tourist_ID'] == tourist_id]['Preferred_Destination'].values[0]}):")
print(recommendations)
```
